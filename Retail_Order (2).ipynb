```python
#Necessary Imports
import numpy as np
import pandas as pd
```

```python
#Read Data From the file 
df=pd.read_csv('orders.csv')
df.head(20)
```

```python
#Study data (data manipulation and handle null values)

df.nunique()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns =df.columns.str.lower()
df.columns =df.columns.str.replace( ' ', '_')
```

```python
df.columns
```

```python
df['ship_mode'].unique()   #similary checked all columns to find row values indicate not data
```

```python
df.replace( ['Not Available' , 'unknown'],np.nan, inplace=True)
```

```python
df['country'].unique()    #only single value
```

```python
df=df.drop(columns=['country'])
```

```python
df['cost_price'].unique()    #has 0 value as cost price
```

```python
df[df['cost_price'] == 0] #also check for df[df['list_price'] == 0]
```

```python
filterdf=df[(df['cost_price'] == 0) & (df['list_price']==0)].index
filterdf
```

```python
df=df.drop(index=filterdf)
```

```python
#derive new columns discount , sale price and profit
df['discount']=df['list_price']*df['discount_percent']*0.01
df['sale_price']=df['list_price']-df['discount']
df['profit']=df['sale_price']-df['cost_price']
df
```

```python
filtered_df=df[(df['discount'] == 0) & (df['sale_price']==0)]
filtered_df
```

```python
filtered_df=df[df['profit'] <= 0]    #since profit can be 0 where sale price is not, more rows are available here
filtered_df           #half of the data have o or less profit
```

```python
#convert order date from object datatype to datetime

df['order_date']=pd.to_datetime(df['order_date'],format="%Y-%m-%d")
```

```python
df.dtypes
```

```python
#drop columns list_price, cost_price,discount_percent 
df.drop(columns=['list_price','cost_price','discount_percent'], inplace=True)
```
